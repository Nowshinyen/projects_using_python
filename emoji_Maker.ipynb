```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Arc, Rectangle
import numpy as np
import random

def draw_face(ax, x=0, y=0, r=1.0, expression="smile"):
    # Face
    face = Circle((x, y), r, facecolor="#FFD54F", edgecolor="black", linewidth=1.4)
    ax.add_patch(face)

    # eye positions
    lx, rx = x - 0.35*r, x + 0.35*r
    ey = y + 0.35*r
    eye_r = 0.10*r

    def eyes_default():
        ax.add_patch(Circle((lx, ey), eye_r, color="black") )
        ax.add_patch(Circle((rx, ey), eye_r, color="black") )

    def eyes_wink():
        ax.add_patch(Rectangle((lx-0.12*r, ey-0.015*r), 0.24*r, 0.03*r, color="black") )
        ax.add_patch(Circle((rx, ey), eye_r, color="black") )

    def eyes_sunglasses() :
        # Lenses
        ax.add_patch(Rectangle((lx-0.12*r, ey-0.10*r), 0.24*r, 0.20*r, color="green"))
        ax.add_patch(Rectangle((rx-0.12*r, ey-0.10*r), 0.24*r, 0.20*r, color="green"))
        # Bridge
        ax.add_patch(Rectangle((x-0.10*r, ey-0.01*r), 0.20*r, 0.02*r, color="black"))

    # Mouth
    def mouth_arc(theta1, theta2, ry=0.55, lw=4):
        arc = Arc((x, y - 0.10*r), 1.2*r, ry*r, theta1=theta1, theta2=theta2, color="black", lw=lw)
        ax.add_patch(arc)

#--------------------------------------------------------------------------------------------------
    if expression == "smile" :
        eyes_default()
        mouth_arc(200, 340)

    elif expression == "wink" :
        eyes_wink()
        mouth_arc(200, 340)

    elif expression == "sunglass" :
        eyes_sunglasses()
        mouth_arc(200, 340)

    ax.set_aspect('equal')
    ax.set_xlim(x - 1.2*r, x + 1.2*r)
    ax.set_ylim(y - 1.2*r, y + 1.2*r)
    ax.axis('off')
```

```python
expression = ["smile","wink","sunglass"]
n= len(expression)
cols = 3
rows = int(np.ceil(n / cols))

fig = plt.figure(figsize=(10, 6))
for i, e in enumerate(expression, 1) :
    ax = plt.subplot(rows, cols, i)
    draw_face(ax, 0, 0, r=1.0, expression=e)
    ax.text(0, -1.45, e, ha="center", va="top", fontsize=10)
plt.tight_layout()
plt.show()
```
