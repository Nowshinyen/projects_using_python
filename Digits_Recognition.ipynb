```python
import matplotlib.pyplot as plt
import random
from sklearn import datasets, metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
digits = datasets.load_digits() #Loads (8×8) images of the numbers 0–9

print("Shape of data:", digits.data.shape) #size of dataset (1797 images, each 64 pixels)
print("Number of target labels:", len(digits.target))
print("Target classes:", set(digits.target))
```

```python
#Shows first 9 digit images

plt.figure(figsize=(6,6))
for a in range(9) :
    plt.subplot(3,3,a+1)
    plt.imshow(digits.images[a], cmap='gray')
    plt.title(f"Label: {digits.target[a]}" )
    plt.axis('off')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    digits.data, digits.target, test_size=0.2, random_state=42) #20% of data is for testing and 80% for training
```

```python
model = LogisticRegression(max_iter=4000)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
accuracy = metrics.accuracy_score(y_test, y_pred)
print("The Accuracy is :", accuracy)
```

```python
cm = metrics.confusion_matrix(y_test, y_pred) #Confusion Matrix
plt.figure(figsize=(6,6))
plt.imshow(cm, cmap='Greens')
plt.title("Confusion Matrix figure")
plt.colorbar()
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

```python
n = random.randint(0, len(X_test)-1 )
plt.imshow(X_test[n].reshape(8,8), cmap='gray')
plt.title(f"Predicted: {model.predict([X_test[n]])[0]}, Actual number : {y_test[n]}")
plt.axis("off")
plt.show()
```
